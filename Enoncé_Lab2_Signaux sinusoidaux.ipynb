## Introduction : Ondes sinusoïdales

Les **signaux sinusoïdaux** sont les éléments de base de la théorie des signaux et des systèmes, à la fois pour des raisons fondamentales (certaines de leurs combinaisons linéaires peuvent utilisées pour reproduire n'importe quelle forme de signaux - *cf* le chapitre sur la décomposition en série de Fourier) et aussi pratiques (technologiques). On utilise en effet presque toujours des signaux sinusoïdaux (électriques ou optiques) lorsqu'on veut transmettre un message sur un canal de communication (le procédé d'adaptation du signal au canal s'appelle la **modulation**), c'est pourquoi on les désigne également sous le vocable d'**ondes sinusoïdales**.

Les signaux sont donc généralement transmis sous forme d'ondes, par exemple d'ondes électromagnétiques, ondes sinusoïdales qui se propagent dans le vide avec une vitesse égale à celle de la lumière, $v=c=3 \ 10^8 m/s$, et dans les milieux matériels transparents, avec une vitesse moindre dépendant de l'indice de réfraction du milieu, selon $v_{Milieu} = \frac{c}{n_{Milieu}}$ ($n_{Milieu}>1)$.  

Dans le domaine du son, une onde sonore sinusoïdale (en abrégé un "son") est une [onde longitudinale](http://blog.soton.ac.uk/soundwaves/files/2013/12/longituddots1.gif) de fréquence unique $f_0$ (dans la gamme $[20Hz,20kHz]$ où l'oreille humaine est sensible) qui se propage dans un milieu, tel que l'air ou l'eau, une vibration qui se propage dans le milieu sous forme de fluctuations périodiques de pression qui résultent du déplacement des molécules qui constituent le milieu.

L'image animée ci-dessous illustre la propagation d'une onde sonore longitudinale émise par une sphère vibrante dans l'air. On y distingue les fronts d'onde marqués par les fluctuations de densité, donc de pression, associées à la propagation de l'onde sonore longitudinale.

## II.1. Onde progressive sinusoïdale : Longueur d'onde (spatiale) et période (temporelle)

Une onde progressive se déplaçant dans un milieu (exemple : le déplacement transverse d'une corde vibrante, ou une onde sonore longitudinale se propageant dans l'air) peut être décrite (modélisée), le long d'un axe donnant la coordonnée x, par une quantité $y$ (par exemple la pression acoustique) dont la valeur dépend de la coordonnée $x$ et du temps $t$, d'expression analytique suivante (pour une direction de propagation de l'onde selon les $x$ croissants) :

$$ y(x,t) = A \ sin(k\ x - \omega \ t ) $$

dont les paramètres sont les grandeurs suivantes : 

- **dans le domaine temporel** : trois grandeurs indépendantes du milieu de propagation de l'onde :
 - $\omega$, la **pulsation** de l'onde, reliée à la **période** (temporelle) $T$ de l'onde par $\omega = \frac{2\pi}{T}$, et à la **fréquence** de l'onde $\nu=\frac{1}{T}=\frac{2\pi}{\omega}$, 
- **dans le domaine spatial** : où les grandeurs dépendent de la **vitesse de propagation $v$ de l'onde** dans le milieu :
 - k, le **vecteur d'onde**, relié à 
 - la longueur d'onde $\lambda$, reliée au vecteur d'onde $k$ par $k=\frac{2\pi}{\lambda}$,

## II.2 Les signaux sinusoïdaux : Génération de sons musicaux

### II.2.A Génération d'un son musical pur

Contrairement aux accords, un son sinusoïdal pur est composé d'une seule composante fréquentielle. Par exemple, le son correspondant à la note LA (A en notation anglaise) est une onde sonore de fréquence $f_0=440 Hz$, créée par un diapason, un haut-parleur, la corde d'un violon ou un autre instrument ou dispositif qui fait osciller l'air à la fréquence donnée. 

Sur un ordinateur, les sons ne sont rien d'autre qu'une suite de chiffres de précision limitée, qu'on peut assimiler à des nombres entiers. Un son donné peut être décrit mathématiquement par un signal analogique sinusoïdal (en abrégé, une sinusoïde) d'expression analytique :

$s(t)=A \ \cos(2πft),$

où A est l'amplitude, f est la fréquence (440 Hz pour la note LA de la gamme de Do) et t, la variable indépendante de cette équation, est le temps. 

Dans un ordinateur, $s(t)$ est évalué à des points discrets dans le temps, déterminés par la fréquence d'échantillonnage $f_s=\frac{1}{T_s}$. On désigne le signal numérique dérivé de $s(t)$ par $s[n]=s(nT_s)$. .

La fonction python ci-dessous vous permet de créer un son musical pur (une portion de sinusoïde, limitée dans le temps) avec une fréquence, une longueur, une amplitude et une fréquence d'échantillonnage spécifiques. Vous pourrez donc écouter un son, et visualiser son contenu sous forme d'un graphe représentant les valeurs des échantillons successifs en fonction de leur indice ou du temps. 

```python
def tone(frequency=440., length=1., amplitude=1., sampleRate=44100., soundType='int8'):
    """ Returns a sine function representing a tune with a given frequency.
    
    :frequency: float/int. Frequency of the tone.
    :length: float/int. Length of the tone in seconds.
    :amplitude: float/int. Amplitude of the tone.
    :sampleRate: float/int. Sampling frequency.
    :soundType: string. Type of the elements in the returned array.
    :returns: float numpy array. Sine function representing the tone.
    """
    t = np.linspace(0,length,math.floor(length*sampleRate))
    data = amplitude*np.sin(2*np.pi*frequency*t)
    return data.astype(soundType)

# Parameters that are being used in the start of this notebook
sampleRate = 44100     # La fréquence d'échantillonnage, celle d'un CD audio ordinaire
                       #
sampwidth = 1          # In bytes. 1 for 8 bit, 2 for 16 bit and 4 for 32 bit
                       #
volumePercent = 50     # Volume percentage
nchannels = 1          # Mono. Only mono works for this notebook

# Some dependent variables
shift = 128 if sampwidth == 1 else 0 # The shift of the 8 bit samples, as explained in the section above.
soundType = 'i' + str(sampwidth)
amplitude = np.iinfo(soundType).min*volumePercent/100.

# Parameters for a A tone lasting 1 second at a sample
frequency = 440
length = 2

# Calculate the sine function for the given parameters
data = tone(frequency, length, amplitude, sampleRate, soundType)
print('data =',data)
print('data max =',max(data))

# Plot the function
plt.plot(data,'.')
plt.xlim([0,1000])
plt.title('Visualization of the A tone')
plt.xlabel('Sample number')
plt.show()

# Open a new .wav-file in "write" mode, set file parameters
# and write to file
data += shift # Shift the samplings if we use 8 bit
with wave.open('Atone.wav', 'w') as file:
    file.setparams((nchannels, sampwidth, sampleRate, 0, 'NONE', ''))
    file.writeframes(data)

# Display the sound file
IPython.display.Audio('Atone.wav')
```

## II.3 Somme de sinusoïdes déphasées : Interférences

Nous allons maintenant nous intéresser à la somme de deux sinusoïdes de même fréquence, qui ne diffèrent que par une différence de phase (un déphasage) $\Delta \phi$.

> **Q3.1** : Prouvez que la somme $s(t)$ des deux signaux sinusoïdaux de fréquence $f_0$ désignés par $a(t)= \sin(2\pi f_0t + \phi_a)$ et $b(t)= \sin(2\pi f_0t + \phi_b)$ est un toujours un signal sinusoïdal, de même fréquence $f_0$, et trouvez l'expression de l'amplitude et de la phase de $s(t)$. **Indication** : On rappelle que $\sin a + \sin b = 2 \sin(\frac{a+b}{2}) \cos(\frac{a-b}{2})$ (*cf* remarque 2 ci-dessous).

> **Q3.1-R** (votre réponse ici) :

**Remarque 1** : La somme de deux signaux de même fréquence mais d'amplitudes différentes est plus compliquée à démontrer ; nous y reviendrons dans la section 2, où la somme de deux sinusoïdes de même fréquence sera plus facile à déduire à partir des sinusoïdes exprimées comme des vecteurs tournants grâce aux exponentielle complexes. 

**Remarque 2** : La démonstration de cette formule de la trigonométrie est possible par l'intermédiaire des nombres complexes, en appliquant la formule d'Euler $\sin \theta = \frac{e^{j\theta}-e^{-j\theta}}{2j}$ (*cf* Section 2 du cours).

$$
\begin{align}
\sin a + \sin b &= \frac{1}{2j} \left( e^{ja}-e^{-ja} \right) + \frac{1}{2j} \left( e^{jb}-e^{-jb} \right) \\
&= \frac{1}{2j} e^{-jb} \left( e^{j(a+b)}-e^{-j(a-b)} \right) + \frac{1}{2j} e^{-ja} \left( e^{j(a+b)}-e^{j(a-b)} \right) \\
&= \frac{1}{2j} \left( e^{j(a+b)} \left( e^{-jb} + e^{-ja} \right) \right) - \frac{1}{2j} \left( e^{-ja} + e^{-jb} \right) \\ 
&= \frac{1}{2j} \left( e^{-jb} + e^{-ja}  \right)  \left( e^{j(a+b)} -1 \right) \\
&= \frac{1}{2j} \left( e^{-jb} + e^{-ja}  \right) e^{j\frac{a+b}{2}} \left( e^{j\frac{a+b}{2}} - e^{j\frac{a+b}{2}} \right) \\
&= \frac{1}{2j} \left(  e^{j\frac{a-b}{2}} +  e^{-j\frac{a-b}{2}}  \right)  \left( e^{j\frac{a+b}{2}} - e^{j\frac{a+b}{2}} \right) \\
&= 2 \frac{e^{j\frac{a-b}{2}} +  e^{-j\frac{a-b}{2}} }{2}  \frac {e^{j\frac{a+b}{2}} - e^{j\frac{a+b}{2}}}{2j} \\
&= 2 \sin(\frac{a+b}{2}) \cos(\frac{a-b}{2})
\end{align}
$$

## II.4 Somme de deux sinusoïdes de fréquences multiples l'une de l'autre

La somme de deux sinusoïdes de fréquences harmoniques (*i.e.* multiples l'une de l'autre, soit $\frac{f_1}{f_0}=n \in \mathbb{N}$) est toujours une sinusoïde de fréquence $f_0$. Vérifiez cette propriété à l'aide de la simulation ci-dessous, où vous pouvez également choisir le déphasage entre les deux sinusoïdes (entre 0 et $2\pi$ par intervalle de $\frac{\pi}{8}$. 

Cette propriété est à la base de la décomposition en série de Fourier, qui sera vue dans la section 2 de ce cours. La fréquence $f_0$ est appelée **fréquence fondamentale**, ses fréquences multiples les **harmoniques** (de rang $n$).


```python
def somme1(fm1,deltaphi):
    A=1
    f0=1/(2*np.pi)
    f1=fm1*f0
    N=2**9
    tt=np.linspace(-1/f0,1/f0,N)
    dt=tt[1]-tt[0]
    x=np.sin(2*np.pi*f0*tt)
    y=np.sin(2*np.pi*f1*tt+deltaphi)
    z=x+y

    plt.plot(tt,x)
    plt.plot(tt,y)
    plt.plot(tt,z,'r',linewidth=3)
    plt.plot(0,0,'ob')
    plt.ylim(-2,2)

    plt.grid('On')
    plt.title('Somme de 2 sinusoïdes harmoniques')
    plt.xlabel('temps (s)')
    plt.ylabel('signaux')
    plt.legend(['x(t)','y(t)','x(t)+y(t)'])
    
    print('f1 est égale à %1d fois f0' %fm1)    #  Formatted output !
    dephi=deltaphi/np.pi
    print('Le déphasage est égal à %1.2f pi' %dephi) 


widgets.interact(somme1, fm1=widgets.IntSlider(min=0, max=10, value=2, description="f1=f0 fois"), 
                deltaphi=widgets.FloatSlider(min=0, max=2*np.pi, value=0, step=np.pi/8, description="Déphasage"))
                 
```

## II.5 Somme de deux sinusoïdes de fréquences proches : Battements

Les musiciens de l'orchestre accordent leur instrument en écoutant les battements entre un signal sonore sinusoïdal $x(t)$ produit par leur instrument et un signal sinusoïdal $y(t)$ (un son pur) produit par une source émettant à la fréquence de référence. 

$$ \begin{align}
x(t) + y(t) &= \ \sin(2\pi f_0t ) + \ \sin(2\pi f_1t + \phi) \\
&= 2 \sin \left( \frac{2 \pi (f_0 + f_1) t + \phi}{2} \right) \cos \left( \frac{2 \pi (f_0-f_1) - \phi}{2} \right) \\
&= A(t) \sin \left( \frac{2 \pi (f_0 + f_1) t + \phi}{2} \right) 
\end{align}
$$

La somme de ces sinusoïdes de fréquences $f_0$ et $f_1$ (la fréquence de référence) aboutit à un produit d'une sinusoïde à une fréquence élevée $f_+=f_0+f_1$ multipliée par une cosinusoïde de fréquence $f_-f_0-f_1$, d'autant plus faible que $f_0$ est proche de $f_1$. Ces deux composantes sont bien visibles sur la figure ci-dessous, où l'on voit que la cosinusoïde de fréquence faible $f_-$ agit comme une enveloppe $A(t)$ qui module l'amplitude instantanée de la sinusoïde de fréquence $f_+$. 

![](https://raw.githubusercontent.com/Pango01/Signal00/main/Beating.png)

Dans la cellule ci-dessous, programmez et visualisez la somme des signaux $x(t)$ (de fréquence $f_0$) et $y(t)$ de fréquence $f_1$. Utilisez un slider pour fixer dynamiquement la valeur de $f_1$ dans la figure.  


```python
N=12800
T0=5
   
t=np.arange(0,T0,T0/N)

te=T0/N
fe=1/te
f0max=fe/2

print('Sampling frequency =',fe)
print('Nyquist frequency =',fe/2)

A=1
# high frequency for sound hearing
f0=180
 
xy=np.zeros(N)
print('Frequency f0 =',f0)

def fa(f1):
    global xy
    x=A*np.cos(2*np.pi*f0*t)
    y=A*np.cos(2*np.pi*f1*t)
    xy=x+y
    plt.plot(t,x,'b',t,y,'g',t,xy,'r')
    plt.axis([0,4*25/f0,-2,2])
    plt.savefig('Beating.png')
    plt.show()
    return f1,xy

widgets.interact(fa, f1=f0+4);
f0W=widgets.IntSlider(min=f0*0.9, max=f0*1.1, step=1, value=f0, continuous_update=False)
widgets.interact(f1=f0W)

# Audio(xy,rate=fe)
# A placer dans une cellule suivante, sinon il joue f1=f0
```

### Accordage d'un instrument de l'orchestre

Les musiciens de l'orchestre accordent leur instrument en écoutant les battements entre un signal sonore sinusoïdal $x(t)$ produit par leur instrument et un signal sinusoïdal $y(t)$ (un son pur) produit par une source émettant à la fréquence de référence. Le script ci-dessous illustre de manière sonore la diminution de la fréquence des battements $f_0-f_1$ lorsque la fréquence $f_0$ émise par l'instrument à accorder s'approche de la fréquence de référence $f_1$.

```python
N=12800
T0=4
   
t=np.arange(0,T0,T0/N)
te=T0/N
fe=1/te
f0max=fe/2

print('Sampling frequency =',fe)
print('Nyquist frequency =',fe/2)

A=0.5
f0=180
ff1=np.arange(f0+6,f0-1,-1)
# xy=np.zeros(np.size(t))
xy=np.zeros(2)
x=A*np.cos(2*np.pi*f0*t)

for f1 in ff1:
    y=A*np.cos(2*np.pi*f1*t)
    z=x+y
    xy=np.concatenate((xy,z))

plt.plot(xy,'r')
# plt.axis([0,4*25/f0,-2,2])
plt.savefig('Tuning.png')
plt.show()

Audio(xy,rate=fe)
```
